```python
# X’s brother is equivalent to X’s father’s or mother’s son not identical to X.
# X’s grandfather is equivalent to X’s father’s or mother’s father.
# X’s grandson is equivalent to X’s son’s or daughter’s son.
# X’s uncle is equivalent to X’s father’s or mother’s brother.
# X’s nephew is equivalent to X’s brother’s or sister’s son.
# Similar rules apply to sister, grandmother, granddaughter, aunt and niece.

# The degree of kinship is defined as follows:
# The distance from X to X’s father, X’s mother, X’s son or X’s daughter is one.
# The distance from X to X’s husband or X’s wife is zero.
# The degree of kinship between X and Y is equal to the shortest distance from X to Y deduced from the above rules.


def degreesOfKinship(inputRelations):
    # Create a dictionary to hold the values of various relationships.
    primitiveKinshipValues = {
      "husband": 0,
      "wife": 0,
      "father": 1,
      "mother": 1,
      "son": 1,
      "daughter": 1,
      "brother": 2,
      "sister": 2,
    }
    
    # Create sets for removing if the previous one is father, son, husband, or brother.
    maleRemoveCurrentSet = {
        ["father", "son"],
        ["mother", "son"],
        ["son", "father"],
        ["daughter", "father"]}
    maleRemovePreviousSet = {"father", "son", "husband", "brother"}
    
    # Create sets for removing if the previous one is mother, daughter, wife, or son.
    femaleRemoveCurrentSet = {
        ["father", "daughter"],
        ["mother", "daughter"],
        ["son", "mother"],
        ["daughter", "mother"]}
    femaleRemovePreviousSet = {"mother", "daughter", "wife", "son"}
    
    # Loop through each line of relations.
    for indexOuter, line in enumerate(inputRelations):       
    # Declare a maximum and minimum degree(s) of kinship.

    minDegrees = 0
    maxDegrees = 0

    # Otherwise, iterate through the currentLine and count the max degrees.
    for indexInner, term in enumerate(line):
        # Check if the term is a list.
        if isinstance(term, list):
            # Assume the first option.
            term1 = term[0][0]
            term2 = term[0][1]
            maxDegrees +=  primitiveKinshipValues[term1] + primitiveKinshipValues[term2]

            # Assume the second option.
            term1 = term[1][0]
            term2 = term[1][1]
            maxDegrees +=  primitiveKinshipValues[term1] + primitiveKinshipValues[term2]
        else:
            # Primitive kinship values are all the same.
            maxDegrees +=  primitiveKinshipValues[term]
        
    # Loop through each line of relations.
    for indexOuter, line in enumerate(inputRelations):       
    # Declare a maximum and minimum degree(s) of kinship.

    minDegrees = 0
    maxDegrees = 0

    #print(line)

    # Otherwise, iterate through the currentLine and count the min degrees.
    for indexInner, term in enumerate(line):
        # Apply the reduction rules.
        # Check if the combinations of each term can be removed or not.
        if (line[indexInner] + line[indexInner+1]).split() in maleRemovePreviousSet:
            # This is the minimum route, so delete the two indices.
            del line[indexInner]
            del line[indexInner+1]

        # Check if the term is a list.
        if isinstance(line[indexInner], list):
            # Assume the first option.
            term1 = term[0][0]
            term2 = term[0][1]
            minDegrees +=  primitiveKinshipValues[term1] + primitiveKinshipValues[term2]

            # Assume the second option.
            term1 = term[1][0]
            term2 = term[1][1]
            minDegrees +=  primitiveKinshipValues[term1] + primitiveKinshipValues[term2]

        else:
            # Primitive kinship values are all the same.
            minDegrees +=  primitiveKinshipValues[term]

        
        # Print the min/max degrees of kinship.
        print(str(maxDegrees) + " " + str(minDegrees))

# degreesOfKinship(relationList)

# Test Input
numberOfRelationLines = int(input("Please input the number of relations: "))

# Create a list to hold each string of adjacency lists.
listOfLines = []

# Create a current vertice variable.
currentVertice = 1

# Create a dictionary for converting to primitive relations.
convertToPrimitiveRelations = {
      "grandfather": [["father", "father"],["mother", "father"]],
      "grandmother": [["father", "mother"],["mother", "mother"]],
      "grandson": [["son", "son"],["daughter", "son"]],
      "granddaughter": [["son", "daughter"],["daughter", "daughter"]],
      "uncle": [["father", "brother"],["mother", "brother"]],
      "aunt": [["father", "sister"],["mother", "sister"]],
      "nephew": [["brother", "son"],["sister", "son"]],
      "niece": [["brother", "daughter"],["sister", "daughter"]]
    }

indexOfListOfLines = 0

# Ask the user to input the adjacency lists.
while indexOfListOfLines < numberOfRelationLines:
    # Turn the input relationsString into a list and remove the "C is A’s ".
    relationStringAsList = (input()).split("’s ")[1:]
    
    # Convert the relations in relationStringAsList to primitive relations.
    for indexOfRelationsLine, relation in enumerate(relationStringAsList):
        if relation in convertToPrimitiveRelations.keys():
            relationStringAsList[indexOfRelationsLine] = convertToPrimitiveRelations[relation]
    # Append the relationStringAsList to the relationsList.
    listOfLines.append(relationStringAsList)
    
    indexOfListOfLines += 1

print(relationStringAsList)

# Test Call of Function
degreesOfKinship(listOfLines)

# Output should be:
#   5 3
#   5 1
#   2 2
#   6 0
#   2 0
#   1 1
#   0 0
```
