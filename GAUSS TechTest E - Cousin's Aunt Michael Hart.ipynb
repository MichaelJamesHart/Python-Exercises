```python
# X’s brother is equivalent to X’s father’s or mother’s son not identical to X.
# X’s grandfather is equivalent to X’s father’s or mother’s father.
# X’s grandson is equivalent to X’s son’s or daughter’s son.
# X’s uncle is equivalent to X’s father’s or mother’s brother.
# X’s nephew is equivalent to X’s brother’s or sister’s son.
# Similar rules apply to sister, grandmother, granddaughter, aunt and niece.

# The degree of kinship is defined as follows:
# The distance from X to X’s father, X’s mother, X’s son or X’s daughter is one.
# The distance from X to X’s husband or X’s wife is zero.
# The degree of kinship between X and Y is equal to the shortest distance from X to Y deduced from the above rules.


def degreesOfKinship(inputRelations):
    # Create a dictionary to hold the values of various relationships.
    kinshipValues = {
      "father": 1,
      "mother": 1,
      "son": 1,
      "daughter": 1,
      
        
      "husband": 0,
      "wife": 0,
      "C is A": 0,
        
      "brother": 1,
      "sister": 1,
      "grandfather": 1,
      "grandmother": 1,
      "grandson": 1,
      "granddaughter": 1,
      "uncle": 2,
      "aunt": 2,
      "nephew": 2,
      "niece": 2
    }
    
    # Remove the first line and set the number as the list length.
    listLength = int(inputRelations.pop(0))
    
    for line in inputRelations:
        # Replace relationships.
        # line = line.replace("brother", "mother’s son")
        # line = line.replace("sister", "mother’s daughter")
        # line = line.replace("grandfather", "mother’s father")
        # line = line.replace("grandmother", "mother’s mother")
        # line = line.replace("uncle", "mother’s brother")
        # line = line.replace("aunt", "mother’s sister")
        # line = line.replace("nephew", "sister’s son")
        # line = line.replace("niece", "sister’s daughter")
        
        # Split the current line based on the position of the apostrophe + s.
        currentLine = line.split("’s ")
        
        # Declare a maximum and minimum degree(s) of kinship.
        minDegrees = 0
        maxDegrees = 0
        
        print(currentLine)
        
        # Otherwise, iterate through the currentLine and count the min/max degrees.
        for term in currentLine:
            maxDegrees +=  kinshipValues[term]
            minDegrees +=  kinshipValues[term]
        
        # Print the min/max degrees of kinship.
        print(str(maxDegrees) + " " + str(minDegrees))

# Declare a sample relation list to test the function.
relationList = [
"7",
"C is A’s father’s brother’s son’s aunt",
"C is A’s mother’s brother’s son’s aunt",
"C is A’s son’s mother’s mother’s son",
"C is A’s aunt’s niece’s aunt’s niece",
"C is A’s father’s son’s brother",
"C is A’s son’s son’s mother",
"C is A"
]

degreesOfKinship(relationList)

# Output should be:
#   5 3
#   5 1
#   2 2
#   6 0
#   2 0
#   1 1
#   0 0
```
